### Seq2Seq Model Training on Romanization Word to Khmer Word

```python
import numpy as np
import pandas as pd
from keras.models import Model
from keras.layers import Input, LSTM, Dense
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import Levenshtein
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction

# Load the datasets
data_rom = pd.read_csv("csv/data_rom_500.csv", header=None)  # Romanized text (input)
data_kh = pd.read_csv("csv/data_kh_500.csv", header=None)    # Khmer text (target)

batch_size = 32  # Batch size for training
epochs = 100  # Number of epochs to train for
latent_dim = 256  # Latent dimensionality of the encoding space

# Prepare datasets for Roman-to-Khmer
input_texts = []
target_texts = []
input_characters = set()
target_characters = set()

for input_text in data_rom[0]:  # Romanized text is now the input
    input_text = str(input_text).strip()
    input_texts.append(input_text)
    for char in input_text:
        input_characters.add(char)

for target_text in data_kh[0]:  # Khmer text is now the target
    target_text = '\t' + str(target_text).strip() + '\n'
    target_texts.append(target_text)
    for char in str(target_text):
        target_characters.add(char)

# Sort characters to ensure consistent token indexing
input_characters = sorted(list(input_characters))
target_characters = sorted(list(target_characters))

num_encoder_tokens = len(input_characters)
num_decoder_tokens = len(target_characters)
max_encoder_seq_length = max([len(txt) for txt in input_texts])
max_decoder_seq_length = max([len(txt) for txt in target_texts])

print('Number of samples:', len(input_texts))
print('Number of unique input tokens:', num_encoder_tokens)
print('Number of unique output tokens:', num_decoder_tokens)
print('Max sequence length for inputs:', max_encoder_seq_length)
print('Max sequence length for outputs:', max_decoder_seq_length)

# Create token indices
input_token_index = {char: i for i, char in enumerate(input_characters)}
target_token_index = {char: i for i, char in enumerate(target_characters)}

# Initialize input and output data for the model
encoder_input_data = np.zeros((len(input_texts), max_encoder_seq_length, num_encoder_tokens), dtype="float32")
decoder_input_data = np.zeros((len(input_texts), max_decoder_seq_length, num_decoder_tokens), dtype="float32")
decoder_target_data = np.zeros((len(input_texts), max_decoder_seq_length, num_decoder_tokens), dtype="float32")

# Populate the data arrays
for i, (input_text, target_text) in enumerate(zip(input_texts, target_texts)):
    for t, char in enumerate(input_text):
        encoder_input_data[i, t, input_token_index[char]] = 1.0
    for t, char in enumerate(target_text):
        decoder_input_data[i, t, target_token_index[char]] = 1.0
        if t > 0:
            # Decoder target data is offset by one timestep
            decoder_target_data[i, t - 1, target_token_index[char]] = 1.0

# Define the encoder
encoder_inputs = Input(shape=(None, num_encoder_tokens))
encoder = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder(encoder_inputs)
encoder_states = [state_h, state_c]

# Define the decoder
decoder_inputs = Input(shape=(None, num_decoder_tokens))
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
decoder_dense = Dense(num_decoder_tokens, activation="softmax")
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model that combines encoder and decoder
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer="rmsprop", loss="categorical_crossentropy", metrics=["accuracy"])

# Train the model
history = model.fit([encoder_input_data, decoder_input_data], decoder_target_data, batch_size=batch_size, epochs=epochs, validation_split=0.2)

# Save the model
model.save("s2s_rom_to_khmer_model_analysis.h5")

# Create inference models
encoder_model = Model(encoder_inputs, encoder_states)

decoder_state_input_h = Input(shape=(latent_dim,))
decoder_state_input_c = Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_outputs, state_h, state_c = decoder_lstm(
    decoder_inputs, initial_state=decoder_states_inputs)
decoder_states = [state_h, state_c]
decoder_outputs = decoder_dense(decoder_outputs)

decoder_model = Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs] + decoder_states)

# Reverse-lookup token index to decode sequences back to text
reverse_input_char_index = {i: char for char, i in input_token_index.items()}
reverse_target_char_index = {i: char for char, i in target_token_index.items()}

# Decode sequence function
def decode_sequence(input_seq):
    # Encode the input as state vectors
    states_value = encoder_model.predict(input_seq)

    # Generate empty target sequence of length 1
    target_seq = np.zeros((1, 1, num_decoder_tokens))
    # Start with the start character
    target_seq[0, 0, target_token_index["\t"]] = 1.0

    decoded_sentence = ""
    stop_condition = False
    while not stop_condition:
        # Predict the next token
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value)

        # Get the most probable token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = reverse_target_char_index[sampled_token_index]
        decoded_sentence += sampled_char

        # Stop condition: end of sequence or max length reached
        if sampled_char == "\n" or len(decoded_sentence) > max_decoder_seq_length:
            stop_condition = True

        # Update target sequence and states
        target_seq = np.zeros((1, 1, num_decoder_tokens))
        target_seq[0, 0, sampled_token_index] = 1.0
        states_value = [h, c]

    return decoded_sentence.strip()

# Total and unique words in Khmer and Romanized datasets
total_rom_words = sum(len(str(text).split()) for text in data_rom[0])
unique_rom_words = len(set(word for text in data_rom[0] for word in str(text).split()))

total_kh_words = sum(len(str(text).split()) for text in data_kh[0])
unique_kh_words = len(set(word for text in data_kh[0] for word in str(text).split()))

print(f"Total Romanized words: {total_rom_words}")
print(f"Unique Romanized words: {unique_rom_words}")
print(f"Total Khmer words: {total_kh_words}")
print(f"Unique Khmer words: {unique_kh_words}")

# Plot word density
rom_word_lengths = [len(str(text).split()) for text in data_rom[0]]
kh_word_lengths = [len(str(text).split()) for text in data_kh[0]]

plt.figure(figsize=(12, 6))
sns.kdeplot(rom_word_lengths, label='Romanized Word Lengths', fill=True)
sns.kdeplot(kh_word_lengths, label='Khmer Word Lengths', fill=True, warn_singular=False)
plt.title('Word Length Density')
plt.xlabel('Word Length')
plt.ylabel('Density')
plt.legend()
plt.show()

# Generate word cloud for Romanized text
plt.figure(figsize=(12, 8))
wc_rom = WordCloud(width=600, height=300).generate(' '.join(input_texts))
plt.imshow(wc_rom, interpolation='bilinear')
plt.title('Word Cloud for Romanized Text')
plt.axis('off')
plt.show()

# Generate word cloud for Khmer text
plt.figure(figsize=(12, 8))
wc_kh = WordCloud(width=600, height=300, font_path='fonts/Taprom.ttf').generate(' '.join(target_texts))
plt.imshow(wc_kh, interpolation='bilinear')
plt.title('Word Cloud for Khmer Text')
plt.axis('off')
plt.show()

# Model summary
model.summary()

# Plot loss graph
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot accuracy graph
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Test the model on some inputs and calculate BLEU score
for seq_index in range(10):  # Test on first 10 samples
    input_seq = encoder_input_data[seq_index: seq_index + 1]
    decoded_sentence = decode_sequence(input_seq)

    # Get ground truth and remove start/end tokens
    reference = list(target_texts[seq_index].strip("\t\n"))
    hypothesis = list(decoded_sentence)

    # Calculate BLEU score

    print(f"Romanized word: {input_texts[seq_index]}")
    print(f"Khmer word: {''.join(reference)}")
    print(f"Predicted Khmer: {decoded_sentence}")



# --- ADD THIS AT THE VERY END OF YOUR FILE ---
def calculate_cer(reference, hypothesis):
    """Calculates Character Error Rate (CER)."""
    if len(reference) == 0:
        return 1.0 if len(hypothesis) > 0 else 0.0
    dist = Levenshtein.distance(reference, hypothesis)
    return dist / len(reference)

def calculate_wer(reference, hypothesis):
    """Calculates Word Error Rate (WER) - treated as whole string match here."""
    ref_words = reference.split()
    hyp_words = hypothesis.split()
    if len(ref_words) == 0:
        return 1.0 if len(hyp_words) > 0 else 0.0
    dist = Levenshtein.distance(ref_words, hyp_words)
    return dist / len(ref_words)

# Initialize score counters
total_cer = 0
total_wer = 0
num_samples_to_test = 20  # Let's test 20 samples to get a better average

print(f"{'Input (Roman)':<25} | {'Target (Khmer)':<20} | {'Predicted':<20} | {'CER':<5}")
print("-" * 85)

for seq_index in range(num_samples_to_test):
    # 1. Prepare Input
    input_seq = encoder_input_data[seq_index: seq_index + 1]
    
    # 2. Get Actual Strings (Clean up the \t and \n tokens)
    input_text = input_texts[seq_index]
    target_text = target_texts[seq_index].strip("\t\n")
    
    # 3. Predict
    decoded_sentence = decode_sequence(input_seq).strip("\t\n")
    
    # 4. Calculate Metrics
    cer = calculate_cer(target_text, decoded_sentence)
    wer = calculate_wer(target_text, decoded_sentence)
    
    total_cer += cer
    total_wer += wer
    
    print(f"{input_text:<25} | {target_text:<20} | {decoded_sentence:<20} | {cer:.2f}")

# Calculate Averages
avg_cer = total_cer / num_samples_to_test
avg_wer = total_wer / num_samples_to_test

print("-" * 85)
print(f"Average Character Error Rate (CER): {avg_cer:.4f}")
print(f"Average Word Error Rate (WER):      {avg_wer:.4f}")
print("-" * 85)

# Interpret the result for your report
print("\nINTERPRETATION:")
print(f"A CER of {avg_cer:.4f} means that on average, {avg_cer*100:.2f}% of characters differ from the ground truth.")



# Initialize lists to store all references and predictions for the whole batch
# BLEU works best when calculated over the entire dataset (Corpus BLEU)
all_references = []
all_hypotheses = []

print("Collecting data for BLEU Score calculation...")

# We will test on the same 'num_samples_to_test' as defined in the previous block
# Or you can increase this range to len(input_texts) to test the WHOLE dataset (recommended for final report)
test_range = range(num_samples_to_test) 

for seq_index in test_range:
    # 1. Input
    input_seq = encoder_input_data[seq_index: seq_index + 1]
    
    # 2. Actual Target (Ground Truth)
    # Tokenize into characters: "hello" -> ['h', 'e', 'l', 'l', 'o']
    target_text = target_texts[seq_index].strip("\t\n")
    reference = list(target_text) 
    
    # 3. Prediction
    decoded_sentence = decode_sequence(input_seq).strip("\t\n")
    hypothesis = list(decoded_sentence)
    
    # 4. Append to lists
    # Note: corpus_bleu expects a list of references for each hypothesis (hence the extra [])
    all_references.append([reference])
    all_hypotheses.append(hypothesis)

# Calculate BLEU-4 (Considers up to 4-character overlaps)
# We use a smoothing function because some Khmer words might be shorter than 4 characters
chencherry = SmoothingFunction()
bleu_score = corpus_bleu(all_references, all_hypotheses, smoothing_function=chencherry.method1)

print("-" * 85)
print(f"BLEU Score: {bleu_score:.4f}")
print("-" * 85)

print("\nINTERPRETATION:")
print(f"A BLEU Score of {bleu_score:.4f} (or {bleu_score*100:.2f}%) indicates how close the prediction is to the reference.")
print("1.0 is a perfect match. > 0.6 is usually considered excellent for transliteration.")
```

```python
import numpy as np
import pandas as pd
from keras.models import Model
from keras.layers import Input, LSTM, Dense
import pickle
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction

def decode_sequence(input_seq, encoder_model, decoder_model, num_decoder_tokens, target_token_index, reverse_target_char_index, max_decoder_seq_length):
    states_value = encoder_model.predict(input_seq, verbose=0)
    target_seq = np.zeros((1, 1, num_decoder_tokens))
    target_seq[0, 0, target_token_index['\t']] = 1.
    decoded_sentence = ''
    stop_condition = False
    while not stop_condition:
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value, verbose=0)
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = reverse_target_char_index[sampled_token_index]
        decoded_sentence += sampled_char
        if (sampled_char == '\n' or len(decoded_sentence) > max_decoder_seq_length): stop_condition = True
        target_seq = np.zeros((1, 1, num_decoder_tokens))
        target_seq[0, 0, sampled_token_index] = 1.
        states_value = [h, c]
    return decoded_sentence.strip()

# ==========================================
# PART A: ROMAN -> KHMER
# ==========================================
print("\n--- PART A: ROMAN -> KHMER ---")
with open("rom2kh_vocab.pkl", "rb") as f: vocab = pickle.load(f)
latent_dim = 256

# Reconstruct Model
enc_inputs = Input(shape=(None, vocab["num_encoder_tokens"]))
enc = LSTM(latent_dim, return_state=True)
enc_out, state_h, state_c = enc(enc_inputs)
dec_inputs = Input(shape=(None, vocab["num_decoder_tokens"]))
dec_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
dec_out, _, _ = dec_lstm(dec_inputs, initial_state=[state_h, state_c])
dec_dense = Dense(vocab["num_decoder_tokens"], activation='softmax')
dec_out = dec_dense(dec_out)
model = Model([enc_inputs, dec_inputs], dec_out)
model.load_weights("s2s_rom_to_khmer.h5") # Load weights

# Inference Setup
enc_model = Model(enc_inputs, [state_h, state_c])
dec_state_h = Input(shape=(latent_dim,)); dec_state_c = Input(shape=(latent_dim,))
dec_out, state_h, state_c = dec_lstm(dec_inputs, initial_state=[dec_state_h, dec_state_c])
dec_model = Model([dec_inputs, dec_state_h, dec_state_c], [dec_dense(dec_out), state_h, state_c])
rev_target_index = {i: c for c, i in vocab["target_token_index"].items()}

# Test
df_rom = pd.read_csv("csv/data_rom_500.csv", header=None)
df_kh = pd.read_csv("csv/data_kh_500.csv", header=None)
refs, hyps = [], []

print(f"{'Input':<20} | {'Target':<20} | {'Pred':<20}")
print("-" * 65)
for i in range(10):
    input_text = str(df_rom[0][i]).strip()
    target_text = str(df_kh[0][i]).strip()
    input_seq = np.zeros((1, vocab["max_encoder_seq_length"], vocab["num_encoder_tokens"]), dtype='float32')
    for t, char in enumerate(input_text):
        if char in vocab["input_token_index"]: input_seq[0, t, vocab["input_token_index"][char]] = 1.
    
    pred = decode_sequence(input_seq, enc_model, dec_model, vocab["num_decoder_tokens"], vocab["target_token_index"], rev_target_index, vocab["max_decoder_seq_length"])
    print(f"{input_text:<20} | {target_text:<20} | {pred:<20}")
    refs.append([list(target_text)]); hyps.append(list(pred))

print(f"BLEU Score: {corpus_bleu(refs, hyps, smoothing_function=SmoothingFunction().method1):.4f}")

# ==========================================
# PART B: KHMER -> ROMAN
# ==========================================
print("\n--- PART B: KHMER -> ROMAN ---")
with open("kh2rom_vocab.pkl", "rb") as f: vocab = pickle.load(f)

# Reconstruct Model (Same architecture, new dimensions)
enc_inputs = Input(shape=(None, vocab["num_encoder_tokens"]))
enc = LSTM(latent_dim, return_state=True)
enc_out, state_h, state_c = enc(enc_inputs)
dec_inputs = Input(shape=(None, vocab["num_decoder_tokens"]))
dec_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
dec_out, _, _ = dec_lstm(dec_inputs, initial_state=[state_h, state_c])
dec_dense = Dense(vocab["num_decoder_tokens"], activation='softmax')
dec_out = dec_dense(dec_out)
model = Model([enc_inputs, dec_inputs], dec_out)
model.load_weights("s2s.h5") # Load weights

# Inference Setup
enc_model = Model(enc_inputs, [state_h, state_c])
dec_state_h = Input(shape=(latent_dim,)); dec_state_c = Input(shape=(latent_dim,))
dec_out, state_h, state_c = dec_lstm(dec_inputs, initial_state=[dec_state_h, dec_state_c])
dec_model = Model([dec_inputs, dec_state_h, dec_state_c], [dec_dense(dec_out), state_h, state_c])
rev_target_index = {i: c for c, i in vocab["target_token_index"].items()}

# Test
refs, hyps = [], []
print(f"{'Input':<20} | {'Target':<20} | {'Pred':<20}")
print("-" * 65)
for i in range(10):
    input_text = str(df_kh[0][i]).strip() # Input is Khmer
    target_text = str(df_rom[0][i]).strip() # Target is Roman
    input_seq = np.zeros((1, vocab["max_encoder_seq_length"], vocab["num_encoder_tokens"]), dtype='float32')
    for t, char in enumerate(input_text):
        if char in vocab["input_token_index"]: input_seq[0, t, vocab["input_token_index"][char]] = 1.
    
    pred = decode_sequence(input_seq, enc_model, dec_model, vocab["num_decoder_tokens"], vocab["target_token_index"], rev_target_index, vocab["max_decoder_seq_length"])
    print(f"{input_text:<20} | {target_text:<20} | {pred:<20}")
    refs.append([list(target_text)]); hyps.append(list(pred))

print(f"BLEU Score: {corpus_bleu(refs, hyps, smoothing_function=SmoothingFunction().method1):.4f}")
```
